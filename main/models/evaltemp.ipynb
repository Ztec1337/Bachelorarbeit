```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from os import path
import seaborn as sns
sns.set()
```

```python
filename = 'dataset.h5'
keyname = '20simpleHam1'
keyname = '20simpleHam_noise'
filepath = path.abspath(path.join(path.dirname("evaltemp.ipynb"), "..", "..", f"main/data/{filename}"))
```

```python
# Load a single chunk => much faster
dataset = pd.read_hdf(filepath, keyname)
```

```python
fig = plt.figure(figsize=(10, 6))
n = 1

plt.plot(dataset["spectrum"][n],label= f" Fieldstrength {dataset['aFieldStrength'][n]:.2f}\n Phase 2nd Order {dataset['b'][n]:.4f} \n Phase 3nd Order {dataset['c'][n]:.6f}")
plt.title("Example Absorbtion Spectrum")
plt.legend()
plt.xlabel("time / steps")
plt.ylabel("intensity")
```
