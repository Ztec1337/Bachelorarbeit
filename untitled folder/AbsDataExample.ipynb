### ---Fourier filtering

```python
def super_gauss(xarray, mean, width, order=2, mirror=False):
    if not hasattr(xarray, '__len__'):
        xarray = np.arange(xarray)
    sgauss = np.exp(-(np.absolute(xarray-mean)**order)/(2*width**order))
#     if mirror:
#         sgauss = np.concatenate((sgauss, sgauss[::-1]))
#     sgauss = np.nan_to_num(sgauss)
    return sgauss

def cos2_window(index1, index2, width, size):
    indmax = np.maximum(index1,index2)
    indmin = np.minimum(index1,index2)    
    arr1 = np.sin(np.pi/2/width*np.array(range(width)))**2.
    arr2 = np.ones(indmax-indmin)
    arr3 = np.cos(np.pi/2/width*np.array(range(width)))**2.
    remain = size-indmax+indmin-2*width
    arr4 = np.zeros(remain)
    return np.roll(np.concatenate((arr1,arr2,arr3,arr4)), indmin-width)
```

```python
index = 0
cos_window = cos2_window(550, 1200, 100, len(intensity[index,:]))
spec = intensity[index,:]*cos_window

freq = np.arange(-range_efreq/2,range_efreq/2,delta_efreq)
gauss_window  = super_gauss(xarray=freq, mean=0, width=1.2, order=8)

fft_spec = np.fft.fftshift(np.fft.fft(spec,axis=0))
filtered = gauss_window*fft_spec
filtered_spec = np.fft.ifft(np.fft.ifftshift(filtered), axis=0)
filtered_OD   = np.real(- np.log(spec/filtered_spec))
```
